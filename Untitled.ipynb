```python
import pandas as pd
import random

from faker import Faker
```

```python
fake = Faker("en_IN")

rows = []

for i in range(100):

    rows.append({
        "Customer Name": fake.name(),
        "Phone Number": str(
            random.randint(
                9000000000,
                9999999999
            )
        ),
        "Email Address": fake.email(),
        "Company Name": fake.company(),
        "City Name": fake.city(),
        "Comments": fake.sentence()
    })

df = pd.DataFrame(rows)

df.head()
```

```python
df.to_csv(
    "sample_leads.csv",
    index=False
)

print("CSV Created Successfully")
```

```python
df = pd.read_csv(
    "sample_leads.csv"
)

df.head()
```

```python
CRM_FIELDS = [

    "created_at",

    "name",

    "email",

    "country_code",

    "mobile_without_country_code",

    "company",

    "city",

    "state",

    "country",

    "lead_owner",

    "crm_status",

    "crm_note",

    "data_source",

    "possession_time",

    "description"
]
```

```python
COLUMN_MAPPING = {

    "lead name": "name",
    "customer name": "name",
    "full name": "name",
    "name": "name",

    "email": "email",
    "email address": "email",
    "email id": "email",

    "mobile": "mobile_without_country_code",
    "phone": "mobile_without_country_code",
    "phone number": "mobile_without_country_code",
    "contact number": "mobile_without_country_code",

    "company": "company",
    "company name": "company",

    "city": "city",
    "city name": "city",

    "state": "state",

    "country": "country",

    "remarks": "crm_note",
    "comments": "crm_note",
    "notes": "crm_note",

    "source": "data_source",
    "lead source": "data_source"
}
```

```python
mapping = {}

for col in df.columns:

    normalized = col.lower().strip()

    if normalized in COLUMN_MAPPING:

        mapping[col] = COLUMN_MAPPING[
            normalized
        ]

mapping
```

```python
STATUSES = [

    "GOOD_LEAD_FOLLOW_UP",

    "DID_NOT_CONNECT",

    "BAD_LEAD",

    "SALE_DONE"
]
```

```python
parsed_records = []

skipped_records = []

for _, row in df.iterrows():

    crm_record = {}

    for field in CRM_FIELDS:
        crm_record[field] = ""

    for source_col, crm_field in mapping.items():

        crm_record[crm_field] = str(
            row[source_col]
        )

    crm_record["country_code"] = "+91"

    crm_record["country"] = "India"

    crm_record["crm_status"] = random.choice(
        STATUSES
    )

    email = crm_record["email"]

    mobile = crm_record[
        "mobile_without_country_code"
    ]

    if (
        email.strip() == ""
        and
        mobile.strip() == ""
    ):
        skipped_records.append(
            crm_record
        )
        continue

    parsed_records.append(
        crm_record
    )

print(
    "Imported:",
    len(parsed_records)
)

print(
    "Skipped:",
    len(skipped_records)
)
```

```python
parsed_df = pd.DataFrame(
    parsed_records
)

parsed_df.head()
```

```python
parsed_df.to_csv(
    "groweasy_crm_output.csv",
    index=False
)

print(
    "Output Saved Successfully"
)
```
